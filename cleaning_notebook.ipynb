# Cleaning column names

```python
df.columns = df.columns.str.strip().str.lower()

# Display the DataFrame to verify changes
print(df)
```

```python
df.info()
```

# Convert runtime into intiger

```python

def convert_runtime(value):
    if isinstance(value, str):
        return int(value.replace(" min", "").strip())
    return value

df['runtime'] = df['runtime'].apply(convert_runtime)

print(df)
```

```python
df.info()
```

# Convert column 'gross' to float

```python

def convert_gross(value):
    if isinstance(value, str):
        # Remove any commas and convert to float
        return float(value.replace(",", ""))
    return float(value)


df['gross'] = df['gross'].apply(convert_gross)

print(df)
```

```python
df.info()
```

# Drop null for column 'gross'

```python
# null values in each column 
null_counts = df.isnull().sum()
print("Number of null values in each column before dropping:")
print(null_counts)
df.info()
```

```python
#Drop null values in column 'gross'

df = df.dropna(subset=['gross'])

display(df)

```
