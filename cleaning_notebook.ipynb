# Import dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
df = pd.read_csv('/Users/marina/Desktop/IRON HACK LABS/wk7-ml-project/imdb_top_1000.csv')

```

# Cleaning column names

```python
df.columns = df.columns.str.strip().str.lower()

# Display the DataFrame to verify changes
print(df)
```

# Convert runtime into intiger

```python

def convert_runtime(value):
    if isinstance(value, str):
        return int(value.replace(" min", "").strip())
    return value

df['runtime'] = df['runtime'].apply(convert_runtime)

print(df)
```

```python
df.info()
```

# Convert column 'gross' to float

```python

def convert_gross(value):
    if isinstance(value, str):
        # Remove any commas and convert to float
        return float(value.replace(",", ""))
    return float(value)


df['gross'] = df['gross'].apply(convert_gross)

print(df)
```

```python
df.info()
```

# regroup 'genre' and 'certificate'

```python

genre_new_cat = {
    'Drama': 'Drama',
    'Romance': 'Drama',
    'Biography': 'Drama',
    'History': 'Drama',
    'War': 'Drama',
    'Action': 'Action/Adventure',
    'Adventure': 'Action/Adventure',
    'Comedy': 'Comedy',
    'Sci-Fi': 'Sci-Fi/Fantasy',
    'Fantasy': 'Sci-Fi/Fantasy',
    'Thriller': 'Thriller/Crime/Mystery',
    'Crime': 'Thriller/Crime/Mystery',
    'Mystery': 'Thriller/Crime/Mystery',
    'Horror': 'Thriller/Crime/Mystery',
    'Animation': 'Other',
    'Family': 'Other',
    'Musical': 'Other',
    'Western': 'Other',
    'Film-Noir': 'Other',
    'Music': 'Other',
    'Sport': 'Other'
}

def categorize_genres(genre_list, category_mapping):
    for genre in genre_list:
        if genre in category_mapping:
            return category_mapping[genre]
    return 'Other'

df['genre_cat'] = df['genre'].apply(lambda x: categorize_genres(x.split(', '), genre_new_cat))

#print(df)

# films per cat
category_counts = df['genre_cat'].value_counts()

print("\nCounts of each category:")
print(category_counts)
```

```python
#'family' - parental guidance suggested

#  mapping categories
certificate_mapping = {
    'A': 'all public',
    'UA': 'family',
    'U': 'all public',
    'R': 'over 18',
    'G': 'all public',
    'PG-13': 'family',
    'PG': 'family',
    'Passed': 'family',
    'Approved': 'all public',
    'TV-PG': 'family',
    'U/A': 'family',
    'GP': 'all public'
}


df['certificate_cat'] = df['certificate'].map(certificate_mapping)


certificate_counts = df['certificate_cat'].value_counts()

print("\nCounts of each category:")
print(certificate_counts)
```

# heatmap

```python

numerical_cols = ['runtime', 'imdb_rating', 'meta_score', 'no_of_votes', 'gross']
corr = df[numerical_cols].corr()


mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.show()
```
