```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
df = df.dropna()
cols_to_drop = [
    'poster_link',
    'series_title',
    'overview'
]

df = df.drop(columns=cols_to_drop)
```

```python
def convert_runtime(value):
    if isinstance(value, str):
        return int(value.replace(" min", "").strip())
    return value

df['runtime'] = df['runtime'].apply(convert_runtime)


def convert_gross(value):
    if isinstance(value, str):
        # Remove any commas and convert to float
        return float(value.replace(",", ""))
    return float(value)


df['gross'] = df['gross'].apply(convert_gross)

df.head()
```

```python
genre_new_cat = {
    'Drama': 'Drama',
    'Romance': 'Drama',
    'Biography': 'Drama',
    'History': 'Drama',
    'War': 'Drama',
    'Action': 'Action/Adventure',
    'Adventure': 'Action/Adventure',
    'Comedy': 'Comedy',
    'Sci-Fi': 'Sci-Fi/Fantasy',
    'Fantasy': 'Sci-Fi/Fantasy',
    'Thriller': 'Thriller/Crime/Mystery',
    'Crime': 'Thriller/Crime/Mystery',
    'Mystery': 'Thriller/Crime/Mystery',
    'Horror': 'Thriller/Crime/Mystery',
    'Animation': 'Other',
    'Family': 'Other',
    'Musical': 'Other',
    'Western': 'Other',
    'Film-Noir': 'Other',
    'Music': 'Other',
    'Sport': 'Other'
}

def categorize_genres(genre_list, category_mapping):
    for genre in genre_list:
        if genre in category_mapping:
            return category_mapping[genre]
    return 'Other'

df['genre'] = df['genre'].apply(lambda x: categorize_genres(x.split(', '), genre_new_cat))

# films per cat
category_counts = df['genre'].value_counts()

print("\nCounts of each category:")
category_counts
```

```python
#  mapping categories
certificate_mapping = {
    'A': 'all public',
    'UA': 'family',
    'U': 'all public',
    'R': 'over 18',
    'G': 'all public',
    'PG-13': 'family',
    'PG': 'family',
    'Passed': 'family',
    'Approved': 'all public',
    'TV-PG': 'family',
    'U/A': 'family',
    'GP': 'all public'
}


df['certificate'] = df['certificate'].map(certificate_mapping)


certificate_counts = df['certificate'].value_counts()

print("\nCounts of each category:")
certificate_counts
```

```python
# Categorize counts
def categorize_count(count):
    if  count==1 :
        return 'low'
    elif 2<= count <= 3:
        return 'average'
    else:
        return 'high'


# Calculate value counts
director_counts = df['director'].value_counts().reset_index()
director_counts.columns = ['director', 'count']

# Apply categorization
director_counts['director_category'] = director_counts['count'].apply(categorize_count)

# Merge the director_counts counts back to the original DataFrame
df = df.merge(director_counts[['director', 'director_category']], on='director', how='left')
```

```python
# Calculate value counts
star1_counts = df['star1'].value_counts().reset_index()
star1_counts.columns = ['star1', 'count']

# Apply categorization
star1_counts['star1_category'] = star1_counts['count'].apply(categorize_count)

# Merge the star1_counts counts back to the original DataFrame
df = df.merge(star1_counts[['star1', 'star1_category']], on='star1', how='left')
```

```python
# Calculate value counts
star2_counts = df['star2'].value_counts().reset_index()
star2_counts.columns = ['star2', 'count']

# Apply categorization
star2_counts['star2_category'] = star2_counts['count'].apply(categorize_count)

# Merge the star1_counts counts back to the original DataFrame
df = df.merge(star2_counts[['star2', 'star2_category']], on='star2', how='left')
```

```python
# Calculate value counts
star3_counts = df['star3'].value_counts().reset_index()
star3_counts.columns = ['star3', 'count']

# Apply categorization
star3_counts['star3_category'] = star3_counts['count'].apply(categorize_count)

# Merge the star1_counts counts back to the original DataFrame
df = df.merge(star3_counts[['star3', 'star3_category']], on='star3', how='left')
```

```python
# Calculate value counts
star4_counts = df['star4'].value_counts().reset_index()
star4_counts.columns = ['star4', 'count']

# Apply categorization
star4_counts['star4_category'] = star4_counts['count'].apply(categorize_count)

# Merge the star1_counts counts back to the original DataFrame
df = df.merge(star4_counts[['star4', 'star4_category']], on='star4', how='left')

df.head()
```

```python
# Remove non-numeric entries from 'released_year'
df = df[pd.to_numeric(df['released_year'], errors='coerce').notnull()]

# Convert years to integers
df['released_year'] = df['released_year'].astype(int)

# Function to categorize each year into its respective decade
def categorize_decade(year):
    return str((year // 10) * 10) + "'s"

# Apply the function to create a new column with decade categories
df['decade'] = df['released_year'].apply(categorize_decade)

# Display the DataFrame
df.head()
```

```python
cols_to_drop = [
    'released_year',
    'director',
    'star1',
    'star2',
    'star3',
    'star4'
]

df = df.drop(columns=cols_to_drop)
df.head()
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```

```python

numerical_cols = ['runtime', 'imdb_rating', 'meta_score', 'no_of_votes', 'gross']
corr = df[numerical_cols].corr()


mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.show()
```
